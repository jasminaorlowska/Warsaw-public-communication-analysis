```python
import pandas as pd
import matplotlib.pyplot as plt
from visualisation.common import results_from_8AM, results_from_12AM, avg_speed
data_8am = pd.read_json(results_from_8AM)
data_12am = pd.read_json(results_from_12AM)
```

From 8,12 .ipynb files we can see that the average speed turned out to be very similar in both cases.
Let's see histogram of average buses speed t 8am and 12 am. 

```python
plt.hist(data_8am[avg_speed], bins=20, edgecolor='black', alpha=0.5, label='8am')
plt.hist(data_12am[avg_speed], bins=20, edgecolor='black', alpha=0.5, label='12am')
plt.xlabel('Average Speed')
plt.ylabel('Number of Buses')
plt.title('Histogram of Average Speed')
plt.legend()
plt.show()
```


At 8 a.m., the distribution of bus speeds appeared to be more varied compared to later in the day. It was noticeable that a greater number of buses were moving at slower speeds during this early hour. Such a phenomenon is unsurprising considering the heightened presence of traffic congestion during the morning rush.

In both instances, the majority of buses maintained average speeds within the range of 30 to 40 kilometers per hour. However, by noon, it was evident that a greater proportion of buses surpassed the prescribed speed limits.

Overall, the results are quite similar in both cases.
