```python
import matplotlib.pyplot as plt
from src.analyze.analyze_speeding import *
from src.common.config import *
from visualisation.common import results_from_12AM, results_from_8AM
from visualisation.common import lon, lat
```

```python
data_8am = pd.read_json(results_from_8AM)
data_12pm = pd.read_json(results_from_12AM)

plt.figure(figsize=(10, 6))

plt.scatter(data_8am[lon], data_8am[lat], marker='o', color='red', label='8am')
plt.scatter(data_12pm[lon], data_12pm[lat], marker='o', color='blue', label='12pm')

margin = 0.025
plt.xlim(WARSAW_LON_MIN - margin, WARSAW_LON_MAX + margin)
plt.ylim(WARSAW_LAT_MIN - WARSAW_LAT_MAX + margin)
plt.axis('equal')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Coordinates')
plt.grid(True)
plt.legend()
plt.show()
```

The graph depicts the coordinates of Warsaw, with dots representing the locations of buses that exceeded the speed limit a
a given moment. Interestingly, both at 8 a.m. and at noon, speeding tendencies were evident in specific areas. Moreover,
it's observable and fair to conclude that such speeding tendencies were consistent along certain routes, as indicated by
the alignment of dots into distinct lines, observed both at 8 a.m. and noon.

Furthermore, there's a clear accumulation of speed limit violations in particular zones, once again evident in both 
timeframes.
